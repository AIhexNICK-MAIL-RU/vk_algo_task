# Two sum


function twoSum(nums, target) {
left = 0;
right = len(nums) - 1;
while (left < right) {
sum = nums[left] + nums[right];
if (sum == target) {
return [left, right];
} else if (sum < target) {
left++;
} else {
right--;
}
}

```python
def two_sum(nums: list[int], target: int) -> list[int]:
    # Инициализируем два указателя для двустороннего поиска
    left = 0
    # Правый указатель начинается с последнего элемента массива
    right = len(nums) - 1
    
    # Продолжаем поиск, пока указатели не встретятся
    while left < right:
        # Вычисляем сумму элементов на текущих позициях
        current_sum = nums[left] + nums[right]
        
        # Если сумма совпадает с целевым значением, возвращаем позиции
        if current_sum == target:
            return [left, right]
        # Если сумма меньше целевого значения, двигаем левый указатель вправо
        elif current_sum < target:
            left += 1
        # Если сумма больше целевого значения, двигаем правый указатель влево
        else:
            right -= 1
    
    # Если решение не найдено, возвращаем пустой список
    return []

# Тест 1: Базовый случай с двумя числами
def test_two_sum_basic():
    # Создаём отсортированный массив чисел
    nums = [2, 7, 11, 15]
    # Искомая сумма
    target = 9
    # Вызываем функцию
    result = two_sum(nums, target)
    # Проверяем результат
    assert result == [0, 1], f"Ожидалось [0, 1], получено {result}"
    print("Тест 1 пройден успешно")

# Тест 2: Случай с несколькими парами чисел
def test_two_sum_multiple():
    # Создаём отсортированный массив с несколькими возможными парами
    nums = [1, 3, 5, 7, 9]
    # Искомая сумма
    target = 8
    # Вызываем функцию
    result = two_sum(nums, target)
    # Проверяем результат
    assert result == [0, 3], f"Ожидалось [0, 3], получено {result}"
    print("Тест 2 пройден успешно")

# Запускаем тесты
if __name__ == "__main__":
    test_two_sum_basic()
    test_two_sum_multiple()
```

```python
def reverse_array(arr: list[int]) -> list[int]:
    # Инициализируем левый указатель в начале массива
    left = 0
    # Инициализируем правый указатель в конце массива
    right = len(arr) - 1
    
    # Продолжаем обход, пока указатели не встретятся
    while left < right:
        # Меняем местами элементы на левой и правой позициях
        # Используем множественное присваивание для обмена без дополнительной переменной
        arr[left], arr[right] = arr[right], arr[left]
        # Перемещаем левый указатель вправо
        left += 1
        # Перемещаем правый указатель влево
        right -= 1
    
    # Возвращаем изменённый массив
    return arr

# Тест 1: Базовый случай с нечётным количеством элементов
def test_reverse_array_odd():
    # Создаём массив с нечётным количеством элементов
    arr = [1, 2, 3, 4, 5]
    # Сохраняем исходный массив для сравнения
    expected = arr.copy()
    # Вызываем функцию
    result = reverse_array(arr)
    # Переворачиваем ожидаемый массив для сравнения
    expected.reverse()
    # Проверяем результат
    assert result == expected, f"Ожидалось {expected}, получено {result}"
    print("Тест 1 пройден успешно")

# Тест 2: Базовый случай с чётным количеством элементов
def test_reverse_array_even():
    # Создаём массив с чётным количеством элементов
    arr = [1, 2, 3, 4]
    # Сохраняем исходный массив для сравнения
    expected = arr.copy()
    # Вызываем функцию
    result = reverse_array(arr)
    # Переворачиваем ожидаемый массив для сравнения
    expected.reverse()
    # Проверяем результат
    assert result == expected, f"Ожидалось {expected}, получено {result}"
    print("Тест 2 пройден успешно")

# Запускаем тесты
if __name__ == "__main__":
    test_reverse_array_odd()
    test_reverse_array_even()
```

# 6 Слияние двух отсортированных массивов

Дано два отсортированных массива. Необходимо
написать функцию которая объединит эти два массива
в один отсортированный.

Создаем два указателя i и j. В цикле
сравниваем значения под этими
индексами. В зависимости от
результата сравнения, в
результирующий массив
записываем элемент либо из
первого массива, либо из второго.

function merge_sorted_arrays(arr1, arr2) {
merged_array = []
i = 0; j = 0
while i < len(arr1) and j < len(arr2) {
if arr1[i] < arr2[j] {
merged_array.append(arr1[i])
i++
} else {
merged_array.append(arr2[j])
j++
}
}
// Если в каком-то из массивов остались
// элементы, добавляем их в конец слияния
merged_array.append(arr1[i:])
merged_array.append(arr2[j:])
return merged_array
}

```python
def merge_sorted_arrays(arr1: list[int], arr2: list[int]) -> list[int]:
    # Создаём пустой список для хранения результата
    merged_array = []
    
    # Инициализируем указатели для обоих массивов
    i = 0  # указатель для первого массива
    j = 0  # указатель для второго массива
    
    # Продолжаем сравнение, пока не достигнем конца одного из массивов
    while i < len(arr1) and j < len(arr2):
        # Сравниваем текущие элементы в массивах
        if arr1[i] < arr2[j]:
            # Если элемент из первого массива меньше, добавляем его
            merged_array.append(arr1[i])
            # Перемещаем указатель в первом массиве вперёд
            i += 1
        else:
            # Если элемент из второго массива меньше или равен, добавляем его
            merged_array.append(arr2[j])
            # Перемещаем указатель во втором массиве вперёд
            j += 1
    
    # Добавляем оставшиеся элементы из первого массива, если они есть
    merged_array.extend(arr1[i:])
    # Добавляем оставшиеся элементы из второго массива, если они есть
    merged_array.extend(arr2[j:])
    
    # Возвращаем полученный отсортированный массив
    return merged_array

# Тест 1: Базовый случай с разными длинами массивов
def test_merge_sorted_arrays_basic():
    # Создаём тестовые массивы
    arr1 = [1, 3, 5, 7]
    arr2 = [2, 4, 6, 8, 9]
    # Ожидаемый результат
    expected = [1, 2, 3, 4, 5, 6, 7, 8, 9]
    # Вызываем функцию
    result = merge_sorted_arrays(arr1, arr2)
    # Проверяем результат
    assert result == expected, f"Ожидалось {expected}, получено {result}"
    print("Тест 1 пройден успешно")

# Тест 2: Случай с одинаковыми элементами
def test_merge_sorted_arrays_duplicates():
    # Создаём тестовые массивы с повторяющимися элементами
    arr1 = [1, 2, 2, 3]
    arr2 = [2, 2, 4]
    # Ожидаемый результат
    expected = [1, 2, 2, 2, 2, 3, 4]
    # Вызываем функцию
    result = merge_sorted_arrays(arr1, arr2)
    # Проверяем результат
    assert result == expected, f"Ожидалось {expected}, получено {result}"
    print("Тест 2 пройден успешно")

# Запускаем тесты
if __name__ == "__main__":
    test_merge_sorted_arrays_basic()
    test_merge_sorted_arrays_duplicates()
```

# 7 Слияние двух отсортированных массивов. Без дополнительных аллокаций


Дано два отсортированных массива. Необходимо
написать функцию которая объединит эти два массива
в один отсортированный.
Первый массив имеет размер, равный
результирующиму массиву, значения в конце равны
нулям. Мы не должны создавать третий массив.
https://leetcode.com/problems/merge-sorted-array/description/

function merge(arr1, arr2) {
pointer1 = len(arr1) - len(arr2) - 1
pointer2 = len(arr2) - 1
pointer3 = len(arr1) - 1
while pointer2 >= 0 {
if pointer1 >= 0
and arr1[pointer1] > arr2[pointer2] {
arr1[pointer3] = arr1[pointer1]
pointer1 -= 1
} else {
arr1[pointer3] = arr2[pointer2]
pointer2 -= 1
}
pointer3 -= 1
}
return arr1
}

```python
def merge(arr1: list[int], arr2: list[int]) -> list[int]:
    # Устанавливаем указатели для обоих массивов
    # pointer1 указывает на последний ненулевой элемент в arr1
    pointer1 = len(arr1) - len(arr2) - 1
    # pointer2 указывает на последний элемент во втором массиве
    pointer2 = len(arr2) - 1
    # pointer3 указывает на последнюю позицию в arr1
    pointer3 = len(arr1) - 1
    
    # Продолжаем, пока не обработаем все элементы из второго массива
    while pointer2 >= 0:
        # Если pointer1 не вышел за границы и элемент в arr1 больше
        if pointer1 >= 0 and arr1[pointer1] > arr2[pointer2]:
            # Копируем элемент из arr1 в текущую позицию
            arr1[pointer3] = arr1[pointer1]
            # Перемещаем pointer1 назад
            pointer1 -= 1
        else:
            # Копируем элемент из arr2 в текущую позицию
            arr1[pointer3] = arr2[pointer2]
            # Перемещаем pointer2 назад
            pointer2 -= 1
        # Перемещаем pointer3 назад
        pointer3 -= 1
    
    # Возвращаем изменённый первый массив
    return arr1

# Тест 1: Базовый случай
def test_merge_basic():
    # Создаём первый массив с нулями в конце
    arr1 = [1, 2, 3, 0, 0, 0]
    # Создаём второй массив
    arr2 = [2, 5, 6]
    # Ожидаемый результат
    expected = [1, 2, 2, 3, 5, 6]
    # Вызываем функцию
    result = merge(arr1, arr2)
    # Проверяем результат
    assert result == expected, f"Ожидалось {expected}, получено {result}"
    print("Тест 1 пройден успешно")

# Тест 2: Случай с одинаковыми элементами
def test_merge_duplicates():
    # Создаём первый массив с нулями в конце
    arr1 = [1, 2, 2, 0, 0, 0]
    # Создаём второй массив
    arr2 = [2, 2, 3]
    # Ожидаемый результат
    expected = [1, 2, 2, 2, 2, 3]
    # Вызываем функцию
    result = merge(arr1, arr2)
    # Проверяем результат
    assert result == expected, f"Ожидалось {expected}, получено {result}"
    print("Тест 2 пройден успешно")

# Запускаем тесты
if __name__ == "__main__":
    test_merge_basic()
    test_merge_duplicates()
```

# 8 Сортировка массива из 0 и 1


Дан массив, содержащий только 0 и 1. Напишите функцию,
которая сортирует его так, чтобы все нули оказались в
начале, а все единички - в конце. Решение должно быть
in-place.
function sort_binary_array(arr) {
left, right = 0, len(arr) - 1
while left < right {
if arr[left] == 1{
arr[left], arr[right] = arr[right], arr[left]
right–
} else {
left++
}
}
return arr
}

```python
def sort_binary_array(arr: list[int]) -> list[int]:
    # Инициализируем два указателя
    left = 0  # указатель на начало массива
    right = len(arr) - 1  # указатель на конец массива
    
    # Продолжаем, пока указатели не встретятся
    while left < right:
        # Если в позиции left стоит 1, меняем его с элементом справа
        if arr[left] == 1:
            # Меняем местами элементы в позициях left и right
            arr[left], arr[right] = arr[right], arr[left]
            # Перемещаем right указатель влево
            right -= 1
        else:
            # Если в позиции left стоит 0, перемещаем left указатель вправо
            left += 1
    
    # Возвращаем отсортированный массив
    return arr

# Тест 1: Базовый случай
def test_sort_binary_basic():
    # Создаём тестовый массив
    arr = [1, 0, 1, 0, 1, 0]
    # Ожидаемый результат
    expected = [0, 0, 0, 1, 1, 1]
    # Вызываем функцию
    result = sort_binary_array(arr)
    # Проверяем результат
    assert result == expected, f"Ожидалось {expected}, получено {result}"
    print("Тест 1 пройден успешно")

# Тест 2: Случай с одним нулём
def test_sort_binary_single_zero():
    # Создаём тестовый массив
    arr = [1, 1, 0, 1, 1]
    # Ожидаемый результат
    expected = [0, 1, 1, 1, 1]
    # Вызываем функцию
    result = sort_binary_array(arr)
    # Проверяем результат
    assert result == expected, f"Ожидалось {expected}, получено {result}"
    print("Тест 2 пройден успешно")

# Запускаем тесты
if __name__ == "__main__":
    test_sort_binary_basic()
    test_sort_binary_single_zero()
```

# 9 Задача флага Нидерладндов

Дан массив состоящий из нулей, единиц и двоек.
Необходимо его отсортировать за линейное время

https://leetcode.com/problems/sort-colors/description

function sortColors(nums) {
low = 0;
mid = 0;
high = len(nums) - 1;
while (mid <= high) {
if (nums[mid] == 0) {
nums[low], nums[mid] = nums[mid], nums[low];
low++;
mid++;
} else if (nums[mid] == 1) {
mid++;
} else if (nums[mid] == 2) {
nums[mid], nums[high] = nums[high], nums[mid];
high--;
}
}
}

```python
def sort_colors(nums: list[int]) -> list[int]:
    # Инициализируем три указателя
    low = 0  # указатель для нулей
    mid = 0  # указатель для единиц
    high = len(nums) - 1  # указатель для двоек
    
    # Продолжаем, пока mid не превысит high
    while mid <= high:
        # Если текущий элемент равен 0, меняем его с элементом в позиции low
        if nums[mid] == 0:
            # Меняем местами элементы в позициях mid и low
            nums[low], nums[mid] = nums[mid], nums[low]
            # Перемещаем оба указателя вперёд
            low += 1
            mid += 1
        # Если текущий элемент равен 1, просто перемещаем mid вперёд
        elif nums[mid] == 1:
            mid += 1
        # Если текущий элемент равен 2, меняем его с элементом в позиции high
        else:
            # Меняем местами элементы в позициях mid и high
            nums[mid], nums[high] = nums[high], nums[mid]
            # Перемещаем high назад (mid не двигаем, так как новый элемент нужно проверить)
            high -= 1
    
    # Возвращаем отсортированный массив
    return nums

# Тест 1: Базовый случай
def test_sort_colors_basic():
    # Создаём тестовый массив
    nums = [2, 0, 2, 1, 1, 0]
    # Ожидаемый результат
    expected = [0, 0, 1, 1, 2, 2]
    # Вызываем функцию
    result = sort_colors(nums)
    # Проверяем результат
    assert result == expected, f"Ожидалось {expected}, получено {result}"
    print("Тест 1 пройден успешно")

# Тест 2: Случай с одним нулём
def test_sort_colors_single_zero():
    # Создаём тестовый массив
    nums = [2, 2, 0, 1, 2, 1]
    # Ожидаемый результат
    expected = [0, 1, 1, 2, 2, 2]
    # Вызываем функцию
    result = sort_colors(nums)
    # Проверяем результат
    assert result == expected, f"Ожидалось {expected}, получено {result}"
    print("Тест 2 пройден успешно")

# Запускаем тесты
if __name__ == "__main__":
    test_sort_colors_basic()
    test_sort_colors_single_zero()
```

# 10 Передвинуть четные числа вперед

Также будем использовать два
индекса. Один будет идти по всему
массиву, а второй будет
увеличиваться, только когда мы
встретим четное число.

function evenFirst(arr) {
evenIndex = 0
for i = 0; i < len(arr); i++ {
if arr[i] % 2 == 0 {
swap( arr[i], arr[evenIndex])
evenIndex += 1
}
}
return arr
}

```python
def evenFirst(arr: list[int]) -> list[int]:
    # Инициализируем указатель для чётных чисел
    evenIndex = 0
    
    # Пробегаем по всему массиву
    for i in range(len(arr)):
        # Проверяем, является ли текущий элемент чётным
        if arr[i] % 2 == 0:
            # Меняем местами текущий элемент с элементом в позиции evenIndex
            # Сначала сохраняем элемент в позиции evenIndex
            temp = arr[evenIndex]
            # Затем меняем элементы
            arr[evenIndex] = arr[i]
            arr[i] = temp
            # Перемещаем evenIndex вперёд
            evenIndex += 1
    
    # Возвращаем изменённый массив
    return arr

# Тест 1: Базовый случай
def test_evenFirst_basic():
    # Создаём тестовый массив
    arr = [1, 2, 3, 4, 5, 6]
    # Ожидаемый результат
    expected = [2, 4, 6, 1, 3, 5]
    # Вызываем функцию
    result = evenFirst(arr)
    # Проверяем результат
    assert result == expected, f"Ожидалось {expected}, получено {result}"
    print("Тест 1 пройден успешно")

# Тест 2: Массив с только нечётными числами
def test_evenFirst_odd_only():
    # Создаём тестовый массив
    arr = [1, 3, 5, 7, 9]
    # Ожидаемый результат (массив не изменится)
    expected = [1, 3, 5, 7, 9]
    # Вызываем функцию
    result = evenFirst(arr)
    # Проверяем результат
    assert result == expected, f"Ожидалось {expected}, получено {result}"
    print("Тест 2 пройден успешно")

# Запускаем тесты
if __name__ == "__main__":
    test_evenFirst_basic()
    test_evenFirst_odd_only()
```

# 11 Нули в конец


В школе прошел экзамен по математике.
Несколько человек списали решения и были
замечены. Этим школьникам поставил 0
баллов. Задача: есть массив с оценками,
среди которых есть 0. Необходимо все
оценки, равные нулю перенести в конец
массива, чтобы все такие школьники
оказались в конце списка.

Пример ввода: [0, 0, 1, 0, 3, 12]
Пример вывода: [1, 3, 12, 0, 0, 0]
Пример ввода: [0, 33, 57, 88, 60, 0, 0, 80, 99]
Пример вывода: [33, 57, 88, 60, 80, 99, 0, 0, 0]
Пример ввода: [0, 0, 0, 18, 16, 0, 0, 77, 99]
Пример вывода: [18, 16, 77, 99, 0, 0, 0, 0, 0]

```python
def move_zeros_to_end(grades: list[int]) -> list[int]:
    # Инициализируем два указателя
    non_zero_index = 0  # указатель для ненулевых элементов
    
    # Пробегаем по всему массиву
    for i in range(len(grades)):
        # Если текущий элемент не равен нулю
        if grades[i] != 0:
            # Меняем его с элементом в позиции non_zero_index
            grades[non_zero_index], grades[i] = grades[i], grades[non_zero_index]
            # Перемещаем non_zero_index вперёд
            non_zero_index += 1
    
    # Возвращаем изменённый массив
    return grades

# Тест 1: Базовый случай
def test_move_zeros_basic():
    # Создаём тестовый массив
    grades = [0, 0, 1, 0, 3, 12]
    # Ожидаемый результат
    expected = [1, 3, 12, 0, 0, 0]
    # Вызываем функцию
    result = move_zeros_to_end(grades)
    # Проверяем результат
    assert result == expected, f"Ожидалось {expected}, получено {result}"
    print("Тест 1 пройден успешно")

# Тест 2: Случай с большим количеством нулей
def test_move_zeros_many_zeros():
    # Создаём тестовый массив
    grades = [0, 0, 0, 18, 16, 0, 0, 77, 99]
    # Ожидаемый результат
    expected = [18, 16, 77, 99, 0, 0, 0, 0, 0]
    # Вызываем функцию
    result = move_zeros_to_end(grades)
    # Проверяем результат
    assert result == expected, f"Ожидалось {expected}, получено {result}"
    print("Тест 2 пройден успешно")

# Тест 3: Случай с одним нулём
def test_move_zeros_single_zero():
    # Создаём тестовый массив
    grades = [33, 57, 88, 60, 0, 80, 99]
    # Ожидаемый результат
    expected = [33, 57, 88, 60, 80, 99, 0]
    # Вызываем функцию
    result = move_zeros_to_end(grades)
    # Проверяем результат
    assert result == expected, f"Ожидалось {expected}, получено {result}"
    print("Тест 3 пройден успешно")

# Запускаем тесты
if __name__ == "__main__":
    test_move_zeros_basic()
    test_move_zeros_many_zeros()
    test_move_zeros_single_zero()
```
